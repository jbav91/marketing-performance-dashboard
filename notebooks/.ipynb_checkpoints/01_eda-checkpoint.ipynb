```python
import pandas as pd
import os
from sqlalchemy import create_engine
from dotenv import load_dotenv

# 1. Cargar las credenciales del archivo .env (subimos un nivel con '../')
load_dotenv('../.env')

# 2. Crear la conexión a PostgreSQL
db_url = f"postgresql+psycopg2://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
engine = create_engine(db_url)

# 3. Extraer los datos de la tabla que creaste
query = "SELECT * FROM marketing_data;"
df = pd.read_sql(query, engine)

# 4. Calcular los KPIs de Marketing (usamos fillna(0) por si hay divisiones por cero)
# CTR (Click-Through Rate)
df['ctr'] = (df['clicks'] / df['impressions']).fillna(0)

# CPC (Cost Per Click) - Usamos mark_budget como gasto
df['cpc'] = (df['mark_spent'] / df['clicks']).fillna(0)

# CPA (Cost Per Acquisition) - Usamos orders como conversión final
df['cpa'] = (df['mark_spent'] / df['orders']).fillna(0)

# ROAS (Return on Ad Spend)
df['roas'] = (df['revenue'] / df['mark_spent']).fillna(0)

# Mostrar una vista previa de los cálculos
print(df[['campaign_id', 'c_date', 'ctr', 'cpc', 'cpa', 'roas']].head())

# 5. Guardar los resultados
# Guardar en la carpeta processed como pide la Fase 2
df.to_csv('../data/processed/metrics_clean.csv', index=False)

# Guardar de vuelta en la base de datos en una nueva tabla
df.to_sql('marketing_kpis', engine, if_exists='replace', index=False)
print("\n¡Métricas calculadas y guardadas con éxito!")
```
